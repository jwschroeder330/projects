# 2. What is the population trend for all ages during the same timeframe? Which population groups are notably increasing and decreasing in popularity?

To answer this question, we plot the differences in age across the period 2004-2018 (this is the time period when Google Trends data is available).

```python
# set the first numpy array: pop_2004
pop_2004 = df.iloc[5, 2:103]
```

```python
# set the first numpy array: pop_2018
pop_2018 = df.iloc[18, 2:103]
```

```python
# subtract the values: pop_diff
pop_diff = pop_2018 - pop_2004
```

At first, we plot for all ages at once to get a general feel for the population trend:

```python
plt.bar(pop_diff.index, pop_diff, color='b')
plt.xlabel('Age')
plt.ylabel('Population Change')
plt.show()
```

Though the x-axis is difficult to read, it is clear something is going on for the middle-aged population. 

Digging deeper, we create a loop to efficiently plot all ages in groups of five:

```python
# plot a bar chart for each age category, split into groups of 5
i = range(0,20)
j = 6
for item in i:
    print(item)
    if item == 0:
        p = (j + (item*5))
        plt.title('Age Range {} to {}'.format(item, p))
        plt.bar(pop_diff[:p].index, pop_diff[:p], color='g')
        plt.xlabel('Age')
        plt.ylabel('Population Change')
        plt.show()
    else:
        p2 = (j + (item*5))
        plt.title('Age Range {} to {}'.format(p, p2))
        plt.bar(pop_diff[p:p2].index, pop_diff[p:p2], color='g')
        plt.xlabel('Age')
        plt.ylabel('Population Change')
        plt.show()
        p = (j + (item*5))
```

With this data in-hand, we conclude that the notable increases are occuring for 60-year-olds, and notable decreases are occuring for 40-year-olds.
